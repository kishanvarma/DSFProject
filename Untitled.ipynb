```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('artist_sort.csv')
```

```python
df1.drop(['groupby_lastfm','artist_rank', 'Alternative & Punk', 'Classical', 'Electronica',
       'Jazz', 'Other', 'Pop', 'Rock', 'Soundtrack', 'Traditional', 'Urban','groupby_youtube', 'groupby_5', 
          'groupby_3','rankby_total_count', 'groupby_3_total_rank', 'groupby_5_total_rank',
          'groupby_3_artist', 'groupby_5_artist','groupby_10_artist'
         ],axis = 1,inplace = True)
```

```python
df2  = df1.sort_values('new_total_count',ascending = False)
df2.reset_index(inplace = True)
```

```python
df2[df2['song'] == 'The Ballad Of The Green Berets']
```

```python
df2[df2['song'] == 'Good Vibrations']
```

```python
df2.index = df2.index + 1
```

```python
lastfm_sort = df2.sort_values('lastfm_rank',ascending = True)
lastfm_sort.reset_index(inplace = True)
lastfm_sort.index = lastfm_sort.index  + 1
lastfm_sort.drop('index',axis = 1,inplace = True)
```

```python
lastfm_sort
```

# Sorted Ascending by lastfm rank,then group in 10.
# MSE Around 10
# Accuracy around 17% or so. 

```python
k = 10
lastfm_sort['groupby_10_lastfm'] = k
row = lastfm_sort.shape[0]
print(row)
for i in range(1,k):
    print((1+(row/k)*(i-1)),"  ",(row/k)*i)
    lastfm_sort['groupby_10_lastfm'].iloc[ (lastfm_sort.index >= int((1 + (i-1)*(row/k)))) & (lastfm_sort.index <= int((row/k)*i))] = i
print("end")
```

```python

print("BEGIN")

"""
Index(['groupby_lastfm', 'groupby_youtube', 'year', 'song', 'artist', 'genre',
       'inverse_score', 'number_of_weeks', 'old_popularity', 'artist_rank',
       'lastfm', 'lastfm_rank', 'youtube', 'youtube_rank', 'new_total_count',
       'genre_category', "Alternative & Punk", 'Classical', 'Electronica',
       'Jazz', 'Other', 'Pop', 'Rock', 'Soundtrack', 'Traditional', 'Urban',
       'number_of_songs_by_artist', 'total_number_of_songs_by_artist',
       'genre_Count_from2005'],
      dtype='object')
"""
from sklearn.cross_validation import train_test_split

train_x, test_x, train_y, test_y = train_test_split(lastfm_sort[['old_popularity',
                                                                'number_of_songs_by_artist',
                                                                'genre_category']],
    lastfm_sort['groupby_10_lastfm'], train_size=0.70)


print("*****************************************LR***************************************")

# Logistic Regression
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error
result =  pd.DataFrame()
# fit a logistic regression model to the data
model = LogisticRegression()
model.fit(train_x, train_y)
print(model)
# make predictions
expected = test_y
predicted = model.predict(test_x)
# summarize the fit of the model
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
print("MSE : ",mean_squared_error(expected,predicted))
print("Accuracy : ",metrics.accuracy_score(expected,predicted, normalize=False))
print("Accuracy : ",metrics.accuracy_score(expected,predicted))
result['expected'] = expected
result['LR'] = predicted



print("*****************************************NB***************************************")

# Naive Bayes
from sklearn import metrics
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import mean_squared_error
model = GaussianNB()
model.fit(train_x, train_y)
print(model)
# make predictions
expected = test_y
predicted = model.predict(test_x)
# summarize the fit of the model
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
print("MSE : ",mean_squared_error(expected,predicted))
print("Accuracy : ",metrics.accuracy_score(expected,predicted ,normalize=False))
print("Accuracy : ",metrics.accuracy_score(expected,predicted))
result['NB'] = predicted


print("*****************************************KNN***************************************")

# knn
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import mean_squared_error
model = KNeighborsClassifier()
model.fit(train_x, train_y)
print(model)
# make predictions
expected = test_y
predicted = model.predict(test_x)
# summarize the fit of the model
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
print("MSE : ",mean_squared_error(expected,predicted))
print("Accuracy : ",metrics.accuracy_score(expected,predicted ,normalize=False))
print("Accuracy : ",metrics.accuracy_score(expected,predicted))
result['KNN'] = predicted



print("*****************************************DEcision Tree***************************************")

# DEcision Tree
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import mean_squared_error
model = DecisionTreeClassifier( splitter = 'random', max_leaf_nodes = 10, min_samples_leaf = 10, max_depth= 10)
model.fit(train_x, train_y)
print(model)
# make predictions
expected = test_y
predicted = model.predict(test_x)
# summarize the fit of the model
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
print("MSE : ",mean_squared_error(expected,predicted))
print("Accuracy : ",metrics.accuracy_score(expected,predicted,normalize=False))
print("Accuracy : ",metrics.accuracy_score(expected,predicted))
result['DT'] = predicted
```
