```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
hot100 = pd.read_csv('data/hot100/hot100.csv')
```

```python
hot100['year'] = pd.DatetimeIndex(hot100['date']).year
hot100['month'] = pd.DatetimeIndex(hot100['date']).month
```

```python
hot100.head()
```

```python
df = hot100[(hot100['date'] > '1969-01-01') & (hot100['date'] < '1969-12-31')]
```

```python
df.tail()
```

```python
df.head()
```

```python
df = df.sort_values('position',ascending = True)
```

```python
df.head(55)
```

```python
df1  = df.groupby(['song','artist','position']).count()
```

```python
df1 = df1.reset_index()
```

```python
df2 = df1.sort_values('position',ascending = True).reset_index()
```

```python
alltop = df2.sort_values(['position','year'],ascending = [True,False]).reset_index()
```

```python
alltop.drop('level_0',axis = 1,inplace = True)
```

```python
alltop.drop_duplicates(subset = ['song','artist'],keep = 'first').reset_index(inplace = True)
```

```python
alltop
```
